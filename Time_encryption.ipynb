```python
import pandas as pd
import numpy as np
from pandasql import sqldf

sepsis = pd.read_csv('C:/Users/ysun1/Documents/MIDS Spring 2018/partial_datacsv.csv', dtype = "object")
sepsis.head(50)
```

##### Parse out day, month, year

```python
sepsis['day'] = sepsis['time'].str[:2]
sepsis['mon'] = sepsis['time'].str[2:5]
sepsis['year'] = sepsis['time'].str[5:9]
sepsis.head(15)
```

```python
sepsis.month.unique()

sepsis['year'] = sepsis['year'].map(int)
```

```python
sepsis['month'] = '00'
sepsis['month'][sepsis['mon'] == 'jan'] = '01'
sepsis['month'][sepsis['mon'] == 'feb'] = '02'
sepsis['month'][sepsis['mon'] == 'mar'] = '03'
sepsis['month'][sepsis['mon'] == 'apr'] = '04'
sepsis['month'][sepsis['mon'] == 'may'] = '05'
sepsis['month'][sepsis['mon'] == 'jun'] = '06'
sepsis['month'][sepsis['mon'] == 'jul'] = '07'
sepsis['month'][sepsis['mon'] == 'aug'] = '08'
sepsis['month'][sepsis['mon'] == 'sep'] = '09'
sepsis['month'][sepsis['mon'] == 'oct'] = '10'
sepsis['month'][sepsis['mon'] == 'nov'] = '11'
sepsis['month'][sepsis['mon'] == 'dec'] = '12'
sepsis.month.unique()
```

##### Adding sesaonality, 1 = spring = Mar, Apr, May; 2 = summer = Jun, Jul, Aug; 3 = fall = Sept, Oct, Nov; 4 = winter = Dec, Jan, Feb

```python
sepsis['seasonality'] = 1
sepsis['seasonality'][(sepsis['month'] == 'jun')|(sepsis['month'] == 'jul')|(sepsis['month'] == 'aug')] = 2
sepsis['seasonality'][(sepsis['month'] == 'sep')|(sepsis['month'] == 'oct')|(sepsis['month'] == 'nov')] = 3
sepsis['seasonality'][(sepsis['month'] == 'dec')|(sepsis['month'] == 'jan')|(sepsis['month'] == 'feb')] = 4
sepsis.head(15)
```

```python
sepsis['date'] = sepsis['year'].map(str) + sepsis['month'].map(str) + sepsis['day'].map(str)
sepsis.head(5)
```

##### Day of Year function

```python
def Day_Of_Year(date, format = '%Y%m%d'):
    date = pd.to_datetime(date, format = format)
    new_year_day = pd.Timestamp(year = date.year, month = 1, day = 1)
    return (date - new_year_day).days + 1


sepsis['DOY'] = sepsis.apply(lambda row: Day_Of_Year(row['date']), axis = 1)
sepsis.head(15)
```

##### Ranking year

```python
sepsis['rank_year'] = sepsis['year'].rank(ascending = 1, method = 'dense')
sepsis.head(15)
```
